```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv("Mall_Customers.csv")
```

```python
dataset
```

```python
X = dataset.iloc[:,[3,4]].values
```

```python
X
```

```python
from sklearn.cluster import OPTICS
from sklearn.datasets import make_blobs
from sklearn.metrics import silhouette_score, adjusted_rand_score
from sklearn.metrics import davies_bouldin_score
X, _ = make_blobs(n_samples=300, centers=3, random_state=42)

optics = OPTICS(min_samples=5, xi=0.05, min_cluster_size=0.05)
y_optics = optics.fit_predict(X)
y_optics
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y_optics, cmap='rainbow', edgecolor='k')
plt.title("OPTICS Clustering")
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()
```

```python
# Evaluate clustering
silhouette_avg = silhouette_score(X, y_optics)
ari = adjusted_rand_score(_, y_optics)
db_index = davies_bouldin_score(X, y_optics)
# Evaluation metrics
print("Silhouette Coefficient: {:.2f}".format(silhouette_avg))
print("Adjusted Rand Index: {:.2f}".format(ari))
print("Davies-Bouldin Index: {:.2f}".format(db_index))
```
