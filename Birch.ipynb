```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv("Mall_Customers.csv")
```

```python
dataset
```

```python
X = dataset.iloc[:,[3,4]].values
```

```python
X
```

```python
from sklearn.cluster import Birch
brc = Birch(n_clusters=None)
y_brc = brc.fit_predict(X)
y_brc
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y_brc, cmap='rainbow', edgecolor='k')
plt.title("Birch Clustering")
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()
```

```python
from sklearn.metrics import silhouette_score
from sklearn.metrics import davies_bouldin_score
# Evaluate clustering
silhouette_avg = silhouette_score(X, y_brc)
db_index = davies_bouldin_score(X, y_brc)
# Evaluation metrics
print("Silhouette Coefficient: {:.2f}".format(silhouette_avg))
print("Davies-Bouldin Index: {:.2f}".format(db_index))
```
